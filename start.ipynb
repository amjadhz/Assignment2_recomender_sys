```python
import pandas as pd
```

```python
file_paths = {
    "documentaries": "./data/documentaries.pkl",
    "entertainment": "./data/entertainment.pkl",
    "films": "./data/films.pkl",
    "from_the_archives": "./data/from-the-archives.pkl",
    "history": "./data/history.pkl",
    "lifestyle": "./data/lifestyle.pkl",
    "music": "./data/music.pkl",
    "science_and_nature": "./data/science-and-nature.pkl",
    "signed": "./data/signed.pkl",
    "sports": "./data/sports.pkl",
    "arts": "./data/arts.pkl",
    "cbbc": "./data/cbbc.pkl",
    "comedy": "./data/comedy.pkl"
}

for title, path in file_paths.items():
    
    df = pd.read_pickle(path)
    print(title)
    print(df.columns)
    print(10 * '-- ')
    print('Length: ')
    print(len(df.columns))
    print(df.shape)
    print(20 * '--- --- ')
```

```python

dataframes = {}
for category, path in file_paths.items():
    try:
        df = pd.read_pickle(path)
        dataframes[category] = df
    except Exception as e:
        dataframes[category] = f"Error loading: {e}"

common_columns = [
    "title", "category", "tags", "description", "image", "duration_txt", "duration_sec", 
    "first_broadcast", "synopsis_small", "synopsis_medium", "synopsis_large"
]

merged_df = pd.concat(
    [df[common_columns] for df in dataframes.values() if isinstance(df, pd.DataFrame)],
    ignore_index=True
)

# Save the final dataset
merged_df.to_csv("./data/bbc_recommender_dataset.csv", index=False)

print("Final dataset saved as 'bbc_recommender_dataset.csv'")
```
